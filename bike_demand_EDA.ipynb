### 방법1) apply(lambda x : )

```python
traincp = train.copy()
def mydef(x):  # x == df
    day_type = 0
    if   x['holiday'] == 0 and x['workingday'] == 0 :
        day_type = 0
    elif x['holiday'] == 0 and x['workingday'] == 1 :
        day_type = 1        
    elif x['holiday'] == 1 and x['workingday'] == 0 :        
        day_type = 2
    return day_type
        
    
traincp["day_type1"] = traincp.apply(lambda df : mydef(df), axis=1)
traincp[['holiday','workingday','day_type1']].value_counts().sort_index()
```

```python
traincp = train.copy()
def mydef(hol, work):
    day_type = 0
    if   hol == 0 and work == 0 :
        day_type = 0
    elif hol == 0 and work == 1 :
        day_type = 1        
    elif hol == 1 and work == 0 :        
        day_type = 2
    return day_type
        
# traincp["day_type1"] = traincp.apply(lambda df : mydef(df), axis=1)    
traincp["day_type2"]   = traincp.apply(lambda df : mydef(df['holiday'], df['workingday']), axis=1)
traincp[['holiday','workingday','day_type2']].value_counts().sort_index()
```

### 방법2) np.where(조건식, 참, 거짓)
* decode(deptno, 10,1,    20,2   30,3    4)
* decode(deptno, 10,1,    4)

```python
traincp['day_type3'] = 0
traincp['day_type3'] = np.where( (traincp['holiday']==0) & (traincp['workingday'] == 1),   1,  traincp['day_type3'])
traincp['day_type3'] = np.where( (traincp['holiday']==1) & (traincp['workingday'] == 0),   2,  traincp['day_type3'])
traincp[['holiday','workingday','day_type3']].value_counts().sort_index()
```

### 회귀 분석(o) / 분류 (x)


- 회귀에서는 accuracy_score 계산 안함
- <font color=red><b>MSE, RMSE, MAE</b></font>, RMSLE(로그씌움), MSLE
- MSE**(1/2) = RMSE

* 주피터 LaTeX -> 수학 수식 표현
* (달러) 사이에 공식을 넣기 (달러)

*  $ MSE = \frac{\mathrm{1} }{\mathrm{n}} \sum\nolimits_{i=1}^{n} (Y_i - \hat{Y_i})^{2}$

*  $ MAE  = \frac{ \sum_{i=1}^{n} |y_i - \hat{y}_i| }{N} $

*  $ RMSE = \sqrt{\frac{1}{n}\Sigma_{i=1}^{n}{\Big(Y_i - \hat{Y_i}\Big)^2}}$

*  $ RMSLE = \sqrt{\frac{1}{n}\Sigma_{i=1}^{n}{\Big({log(Y_i+1) - log(\hat{Y_i}+1)}\Big)^2}}$

### MSE(mean_squared_error) 
* mse에 루트를 씌운게 rmse
* 분산 -> <b>편차(RMSE)보다 error에 더 민감함</b>
* outlier에 더 민감함
* mse는 아래가 뾰족한 이차곡선으로 그려져서 순간 기울기를 구하기 해서 미분이 가능
* mse 적분을 통해 아래 넓이(확률값) 또한 구할 수 있음
* 단점 : 오버피팅(과대적합)이 일어날 수 있음 -> mse가 낮다고 좋은게 아닐수 있음

### MAE(mean_absolute_error) : |error|의 평균
- MAE가 일반적으로 RMSE 보다 작음
- MAE 모든 값에 대한 가중치가 없음 -> rmse는 outlier에 대한 가중치를 줄수 있음
- outlier가 없는 경우 가끔 사용할지도?
